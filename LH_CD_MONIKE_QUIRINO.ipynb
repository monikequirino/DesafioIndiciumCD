## Modelo

1. Limpeza dos dados: durante a análise exploratória já fizemos alguns processos de limpeza. Porém, ainda é necessário considerar a remoção de outliers para aumentar a precisão do modelo a ser construído (com base nos boxplots visualizados durante a análise exploratória e nos desvios-padrão observados, é nítido a presença de valores extremos).


```python
Q1 = data_precificacao_aluguel.quantile(0.25, numeric_only=True)
Q3 = data_precificacao_aluguel.quantile(0.75, numeric_only=True)
IQR = Q3 - Q1
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR
data_precificacao_semoutliers = data_precificacao_aluguel[~((data_precificacao_aluguel < (Q1 - IQR)) | (data_precificacao_aluguel > (Q3 + IQR))).any(axis=1)] 
data_precificacao_semoutliers

```

3. Identificação das variáveis (utilizando somente as variáveis numéricas de interesse)

```python
Var_numerica = ['latitude', 'longitude', 'minimo_noites', 'numero_de_reviews', 'reviews_por_mes', 'calculado_host_listings_count', 'disponibilidade_365']

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler

preprocessor = ColumnTransformer(
    transformers=[
        ('numerica', StandardScaler(), Var_numerica)])
```

4. Construção de três modelos (+treinamento e avaliação)

```python
#Regressão linear
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

modelo1 = LinearRegression()

from sklearn.pipeline import Pipeline

pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                           ('modelo1', modelo1)])

pipeline.fit(X_treino, y_treino)

y_predito = pipeline.predict(X_test)
RMSE = np.sqrt(mean_squared_error(y_test, y_predito))
print(f'RMSE: {RMSE}')
```

```python
#Random Forest

from sklearn.ensemble import RandomForestRegressor

modelo2 = RandomForestRegressor()

pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                           ('modelo2', modelo2)])

pipeline.fit(X_treino, y_treino)

y_predito = pipeline.predict(X_test)
RMSE = np.sqrt(mean_squared_error(y_test, y_predito))
print(f'RMSE: {RMSE}')
```

```python
# Gradient Booster

from sklearn.ensemble import GradientBoostingRegressor

modelo3 = GradientBoostingRegressor()

pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                           ('modelo3', modelo3)])

pipeline.fit(X_treino, y_treino)

y_predito = pipeline.predict(X_test)
RMSE = np.sqrt(mean_squared_error(y_test, y_predito))
print(f'RMSE: {RMSE}')
```

Com base no RMSE, o modelo de melhor performance, dentre os três modelos construídos e treinados, é o Modelo de Gradient Booster (menor RMSE, maior precisão).
